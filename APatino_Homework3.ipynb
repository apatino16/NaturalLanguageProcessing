## Problem 3
Suppose a user enters a query to a search engine. The query vector is (4,23).

‚Ä¢ Compute the Cosine distance between the normalized query vector and all the 5 normalized vectors representing documents used in problem #1.

‚Ä¢ Out of the 5 documents used in problem#1, which document is closest to the query vector?

```python
# Cosine distance

# List of all normalized vectors
normalized_vectors = [normalized_A, normalized_B, normalized_C, normalized_D, normalized_E]
vector_names = ['A', 'B', 'C', 'D', 'E']

# Query vector
query = np.array([4, 23])

# Normalize the query vector
normalized_query = query / np.linalg.norm(query)

# Dictionary to hold the Cosine distances
cosine_distances = {}

# Compute Cosine distance between the normalized query vector and each normalized document vector
for i, normalized_vector in enumerate(normalized_vectors):
    # Cosine similarity
    cosine_sim = np.dot(normalized_query, normalized_vector)
    # Cosine distance
    cosine_dist = 1 - cosine_sim
    cosine_distances[vector_names[i]] = cosine_dist

# Find the document closest to the query vector based on Cosine distance
closest_document = min(cosine_distances, key=cosine_distances.get)
closest_distance = cosine_distances[closest_document]


print(cosine_distances, closest_document, closest_distance)

print('Document B is closest to the query vector.')
```

Zipf‚Äôs Law states that the Rank and the Frequency of words in a large corpus has the following relationship.

`ùëÉùëõ ~ 1/ùëõùëé`

The value of ‚Äòa‚Äô varies from 1.0 ‚Äì 1.5. For this assignment, we will assume that the value of ‚Äòa‚Äô = 1.

Show that the words of ‚ÄúHamlet‚Äù and the ‚ÄúRomeo & Juliet‚Äù plays written by William Shakespeare have Zipf‚Äôs distribution.

To prove that a data has Zipf‚Äôs distribution, plot a graph with the following characteristics.

If a plot displays almost a straight line, it proves that the data has Zipf‚Äôs distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import re

def read_and_preprocess(file_path):
    # Read the text file
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().lower()  # Convert to lowercase

    # Remove punctuation using a regular expression
    text = re.sub(r'[\W_]+', ' ', text)

    # Split the text into words
    words = text.split()

    return words

def plot_zipf(word_counts, title):
    # Sort the word counts dictionary by frequency in descending order
    sorted_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

    # Extract ranks and frequencies
    ranks = np.arange(1, len(sorted_counts) + 1)
    frequencies = [freq for _, freq in sorted_counts]

    # Log transformation
    log_ranks = np.log(ranks)
    log_frequencies = np.log(frequencies)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(log_ranks, log_frequencies, marker='o', linestyle='None', markersize=5)
    plt.title(f'Zipf\'s Law - {title}')
    plt.xlabel('Log of Rank')
    plt.ylabel('Log of Frequency')
    plt.grid(True)
    plt.show()


# Read and preprocess the texts
hamlet_words = read_and_preprocess('Hamlet.txt')
romeo_juliet_words = read_and_preprocess('RomeoJuliet.txt')

# Count the frequency of each word
hamlet_word_counts = Counter(hamlet_words)
romeo_juliet_word_counts = Counter(romeo_juliet_words)

# Plot Zipf's Law
plot_zipf(hamlet_word_counts, 'Hamlet')
plot_zipf(romeo_juliet_word_counts, 'Romeo & Juliet')

plt.show()
```

Compute the TF-IDF vectors for the following 2 documents.

Document 1 (converted in small letters)

*disneyland is in orange county*

Document 2 (converted into small letters)

*orange county disneyland is better than orlando disneyland*

Compute the ‚ÄòStandard TF-IDF value for each word in both documents using either Excel or Python.

Next, use the `CountVectorizer` and `TfidfTransformer` packages from `Scikit-Learn` library and compute the values of TF-IDF.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

# Define the documents
corpus = [
    "disneyland is in orange county",
    "orange county disneyland is better than orlando disneyland"
]

# Scikit-Learn's CountVectorizer and TfidfTransformer
# Initialize the CountVectorizer
vect = CountVectorizer()

# Fit and transform the documents
X = vect.fit_transform(corpus)

# Initialize the TfidfTransformer
transformer = TfidfTransformer()

# Fit and transform the count matrix to TF-IDF representation
tfidf = transformer.fit_transform(X)

# Display the TF-IDF values
print("TF-IDF Matrix using Scikit-Learn:")
print(tfidf.toarray())

# Display the feature names (words)
print("\nFeature Names (Words):")
print(vect.get_feature_names_out())
```
